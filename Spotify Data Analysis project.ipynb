```python
tracks.set_index('release_date',inplace=True)
tracks.head()
```

```python
## making durations in seconds and delete ms
```

```python
tracks['duration'] = tracks['duration_ms'].apply(lambda x: round(x/1000))
tracks.drop(columns ='duration_ms',inplace =True, axis = 1)
```

```python
tracks.duration.head()
```

```python
corr_df = tracks.drop(['key','mode','explicit'],axis =1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot=True,fmt='.1g',vmax =1,center =0,cmap='inferno',linewidths=1,linecolor ='Black')
plt.title('Correlation Heatmap between variables',size =20)
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation =90)
```

```python
tracks['dates'] =tracks.index.get_level_values('release_date')
tracks.dates =pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
```

```python
sns.displot(years,discrete =True,aspect =2 ,height =5 ,kind='hist').set(title ='Number of songs per year')
```

```python
plt.figure(figsize=(14,6))
plt.title('Years Vs Duration')
sns.lineplot(x =years ,y =tracks['duration'])
plt.xticks(rotation = 60)
```
